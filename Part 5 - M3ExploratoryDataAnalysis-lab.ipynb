## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
sns.boxplot(df['ConvertedComp'])

plt.title('Distribution Data of Converted Annual USD Salaries')
plt.ylabel('Annual USD Salaries')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
df['ConvertedComp'].describe()
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3-Q1
IQR
```

Find out the upper and lower bounds.


```python
upper_bounds = Q3 + 1.5*IQR
lower_bounds = Q1 - 1.5*IQR
print('Upper bounds:', upper_bounds)
print('Lower bounds:', lower_bounds)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
outlier=[]
for value in df['ConvertedComp']:
    if (value>upper_bounds or value<lower_bounds):
        outlier.append(value)

print('There are', len(outlier), 'outliers in the ConvertedComp column')
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df_new = df[~df['ConvertedComp'].isin(outlier)]
df_new['ConvertedComp'].describe()
```
